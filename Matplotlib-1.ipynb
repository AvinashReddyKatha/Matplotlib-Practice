```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[38,36,35,32,34,30,38]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(x,y,'b+--')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='')
```

```python
plt.plot(x,y,color='blue',marker='+',linestyle='',markersize=20)
```

```python
plt.plot(x,y,'g<',alpha=0.5)
```
